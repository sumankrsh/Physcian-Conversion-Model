```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('example_spark') \
    .getOrCreate()

# Creating a dataframe
data = [
    (1, "Gilbert Gathara", "24", "Nairobi"), 
    (2, "Someone Else", "32", "Nowhere")
]
headers = ("id", "Name", "Age", "Location")
df = spark.createDataFrame(data, headers)

# Display the DataFrame
df.show()
```

```python
df = spark.createDataFrame([
("Alice", 1),
("Bob", 2),
("Charlie", 3),
], ["Name", "Value"])

df.show()
```

```python
# Define your lists
list1 = ["a", "b", "c", "d"]
list2 = [1, 2, 3, 4]

# Create an RDD from the lists and convert it to a DataFrame
rdd = spark.sparkContext.parallelize(list(zip(list1, list2)))
df = rdd.toDF(["Column1", "Column2"])

# Show the DataFrame
df.show()
```
